### Analysis
According to the temps vs latitude maps,  the maximum temps in the Souther Hemisphere (neg latitudes) are all relatively high,( > 40) where as the temps in the northern hemisphere, decrease down to a maximum temperature of -20 degrees for cities that are furthest from the equator.

The humidity vs latitude graph is interesting.   While the humidity of cities located on the the equator all have significant humidity levels (>60%) some of the places with the lowest humidity can be found around the 20 and -20 degree latitudes.   As the latitudes approach 90 and -90, the humidity once again reaches highs of > 50 degrees
.
The cities with the highest windspeeds (>35 mph) all appear to be found in latitudes between 60 and 80 degrees.



```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

print(api_key)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    print(f" city is {city} lat is {lat_lng[0]}, and long is {lat_lng[1]}")
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
print(cities[2])
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"

city_weather = []

for city in cities:
    weather_url = base_url + "appid=" + api_key +"&q=" + city +"&units=imperial"
    response = requests.get(weather_url).json()
    if response["cod"] == 200:
        print(f"Adding weather data for {city}")
        city_weather.append(response)
    else:
        print(f"City named {city} not found. Skipping...")

    
```

```python
print(len(city_weather))
pprint(city_weather[200])
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Building the dataframe

#weatherpd.rename(columns = {'0':'City'}, inplace = True)
#weatherpd.columns["City"]

#Step through each item in weather_cities to get the needed data
cityname = []
clouds = []
country = []
date = []
humidity = []
lat = []
lng = []
maxtemp = []
windspeed = []

for city in city_weather:
    n = city["name"]
    cityname.append(n)
    c = city["clouds"]["all"]
    clouds.append(c)
    ctry = city["sys"]["country"]
    country.append(ctry)
    d = city["dt"]
    date.append(d)
    h = city["main"]["humidity"]
    humidity.append(h)
    la = city["coord"]["lat"]
    lat.append(la)
    lo = city["coord"]["lon"]
    lng.append(lo)
    t = city["main"]["temp_max"]
    maxtemp.append(t)
    w = city["wind"]["speed"]
    windspeed.append(w)

weatherpd = pd.DataFrame(cityname)
#weatherpd.rename(columns = {'0':'City'}, inplace = True)
#weatherpd.columns["City"]
weatherpd["Cloudiness"] = clouds
weatherpd["Country"] = country
weatherpd["Date"] = date
weatherpd["Humidity"] = humidity
weatherpd["Latitude"] = lat
weatherpd["Longitude"] = lng
weatherpd["Max Temp"] = maxtemp
weatherpd["Wind Speed"] = windspeed
weatherpd.head(5)


    
    
    
```

```python
#write dataframe to a csv
weatherpd.to_csv(output_data_file)
```
